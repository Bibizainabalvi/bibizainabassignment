```python
# 4 Write a python program that accepts the radius of a circle from the user and compute the area.

radius = input(f'Type the radius of the circle here! {int()}')
pi = 3.142
area_of_circle = float(pi) * int(radius) ** 2
print(f'Area of the circle is {area_of_circle}')
```

```python
# 5 Write a python program which accepts the user's first and last name and print them in reverse
# order with space between them.

first_name = input(f'Type first name here!{str()}')
last_name = input(f'Type last name here!{str()}')

first = first_name
second = last_name
full_name = f'{second} {first}'
print(full_name)
```
